```python
import time
import requests
import shutil
from concurrent.futures import ThreadPoolExecutor, as_completed
from requests.adapters import HTTPAdapter
from requests.packages.urllib3.util.retry import Retry
```

```python
def download_iiif_image(iiif_url, f_out, session):
    try:
        with session.get(iiif_url, stream=True) as r:
            if r.status_code == 200:
                r.raw.decode_content = True
                with open(f_out, 'wb') as f:
                    shutil.copyfileobj(r.raw, f)
                return f"Downloaded: {iiif_url}"
            else:
                return f"Failed: {iiif_url} with status {r.status_code}"
    except Exception as e:
        return f"Error: {iiif_url} -> {str(e)}"
```

```python
from concurrent.futures import ThreadPoolExecutor, as_completed

def download_images_parallel(image_urls, page_ids, out_dir_images, max_workers=4, delay=0.5):
    session = create_session()

    def download_if_missing(iiif_url, page_id):
        dest_path = os.path.join(out_dir_images, page_id + ".jpg")
        if not os.path.exists(dest_path):
            download_iiif_image(iiif_url, dest_path, session)
            time.sleep(delay)  # Optional delay to spread out requests
            return f"Downloaded: {iiif_url} -> {dest_path}"
        else:
            return f"Skipped: {dest_path}, already exists."

    with ThreadPoolExecutor(max_workers=max_workers) as executor:
        futures = [
            executor.submit(download_if_missing, iiif_url, page_id)
            for iiif_url, page_id in zip(image_urls, page_ids)
        ]
        for future in tqdm(as_completed(futures), total=len(image_urls), desc="Downloading Images"):
            print(future.result())
```

```python
# output dirs
out_dir = "/scratch/students/danae/data"
out_dir_images = os.path.join(out_dir, "images")

os.makedirs(out_dir_images, exist_ok=True)

# BNF only
df = uris_df[uris_df["iiif_img_base_uri"].str.startswith("https://iiif.eluxemburgensia")]
# download images
download_images_parallel(image_urls=df["iiif_img_base_uri"], page_ids=df["page_id"], out_dir_images=out_dir_images, delay=0.5)
```
